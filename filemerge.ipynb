```python
history = pd.read_csv('historyenv.csv')
med = pd.read_csv('evaluationmedical.csv')
survey = pd.read_csv('survey_vitals.csv')
```

```python
m = 0
h = 0
for o in survey.objectId:
    in_med = False
    in_history = False
    for om in med.objectId:
        if om == o:
            in_med = True
    for oh in history.objectId:
        if oh == o:
            in_history = True
    if not in_med:
        m+=1
    if not in_history:
        h+=1
        
print("M: " + str(m))
print("H: " + str(h))
```
