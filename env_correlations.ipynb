```python
import pandas as pd
import numpy as np 
```

```python
env = pd.read_csv("history_environment.csv")
env.head()
```

```python
env.columns
```

```python
env.drop(["objectId", "latitude", "longitude"], axis=1, inplace = True)
```

```python
dummies = pd.get_dummies(env)
dummies.head()
```

```python
dc = dummies.corr()
dc
```

```python
csigned = dummies.corr()


c = dummies.corr().abs()


s = c.unstack()


so = s.sort_values(ascending=False)


so_noindex = so.reset_index()


so_nosame = so_noindex[so_noindex['level_0'] != so_noindex['level_1']]


def swap(x):
    if (x['level_0'] > x['level_1']): 
        temp = x.level_0
        x.level_0 = x.level_1
        x.level_1 = temp
    return x

so_alpha = so_nosame.apply(swap,axis=1)


so_alpha_nodups = so_alpha.drop_duplicates()


so_alpha_nodups_signed = so_alpha_nodups.apply(lambda x: csigned.loc[x.level_0][x.level_1],axis=1)


final = pd.concat([so_alpha_nodups.drop(0,axis=1),so_alpha_nodups_signed],axis=1)
final.columns=['var1','var2','corr']
```

```python
topcorr = final[abs(final['corr']) > 0.3]
topcorr
```
